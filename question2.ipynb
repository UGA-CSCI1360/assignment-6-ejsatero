```python
# You solution goes here
# you are encouraged to use multiple cells in the notebook

import pandas as pd
import matplotlib.pyplot as plt

train_data = pd.read_csv('train.csv')

plt.figure(figsize=(10, 6))

survived_data = train_data[train_data['Survived'] == 1]

plt.scatter(survived_data['Age'], survived_data['Fare'], label='Survivals', marker='o')

casualties_data = train_data[train_data['Survived'] == 0]
plt.scatter(casualties_data['Age'], casualties_data['Fare'], label='Casualties', marker='x')

plt.xlabel('Age')
plt.ylabel('Ticket Fare')
plt.title('Scatter Plot of Age vs. Ticket Fare for Survivals and Casualties')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

survived_data['Embarked'].value_counts().sort_index().plot(kind='bar', color='blue', alpha=0.7, label='Survivals')

plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.title('Histogram of Survivals from Each Embarkation Port')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

casualties_data['Embarked'].value_counts().sort_index().plot(kind='bar', color='red', alpha=0.7, label='Casualties')

plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.title('Histogram of Casualties from Each Embarkation Port')
plt.legend()

plt.show()
```
